# Examples of numerical semigroups with embedding dimension three and minimal catenary degree

```gap
LoadPackage("numericalsgps");;
```

In this notebook, we give some examples of numerical semigroups with embedding dimension three  such that the maximum of their Delta set plus two is equal to their catenary degree. They are performed with the package $\texttt{numericalsgps}$ for $\texttt{GAP}$.

Let $S=\langle n_1$, $n_2$, $n_3\rangle$ with $2<n_1<n_2<n_3$ be a numerical semigroup with embedding dimension three. Given $\{i, j, k\}=\{1,2,3\}$, define
$$c_i=\min\{k\in\mathbb{Z}^+ \mid kn_i \in \langle n_j,n_k \rangle\}.$$
Then there exist some $r_{ij}$, $r_{ik} \in \mathbb{N}$ such that
$$c_in_i=r_{ij}n_j+r_{ik}n_k.\label{eq:2}$$
We know that
$$\textrm{Betti}(S)=\{c_1n_1,c_2n_2,c_3n_3\}$$
and that $S$ is nonsymmetric if and only if $\#\textrm{Betti}(S)=3$.

We distinguish three cases depending on whether $\#\textrm{Betti}(S)$ is one, two or three.

### 1. A single Betti element

If $\#\textrm{Betti}(S)=1$, then $\textrm{max}\Delta(S)+2<\textrm{c}(S)$. Thus, we have no examples.

### 2. Two Betti elements

If $\#\textrm{Betti}(S)=2$, then $S$ is symmetric and it can be written as $$S=\langle am_1, am_2, bm_1+cm_2 \rangle$$ with $m_1<m_2$ two relatively prime integers greater than one and $a$, $b$ and $c$ nonnegative integers with $a\geq 2$, $b+c \geq 2$ and $\gcd(a, bm_1 + cm_2)=1$.

The only cases in which $S$ has minimal catenary degree are the following.

#### 2.1. $a=2$ and $m_2 < b+c-\big\lfloor\frac{b}{m_2}\big\rfloor(m_2-m_1)$
    
For example, if $m_1=5$, $m_2=7$, $b=5$, $c=4$, $a=2$, that is, $S=\langle10,14,53\rangle$, and with the help of $\texttt{numericalsgps}$ we obtain that

```gap
s:=NumericalSemigroup(10,14,53);;
DeltaSet(s);
CatenaryDegree(s);
```

#### 2.2. $a=2<m_1$ and $m_2 = b+c-\big\lfloor\frac{b}{m_2}\big\rfloor(m_2-m_1)$
    
For example, if $m_1=3$, $m_2=4$, $b=1$, $c=3$, $a=2$, that is, $S=\langle6,8,15\rangle$, then

```gap
s:=NumericalSemigroup(6,8,15);;
DeltaSet(s);
CatenaryDegree(s);
```

Let us have a look at the factorization graphs of the Betti elements of the semigroup

```gap
BettiElements(s);
```

```gap
JupyterSplashDot(DotFactorizationGraph(Factorizations(30,s)));
```

For example, if $m_1=2$, $m_2=5$, $b=1$, $c=3$, $a=2$, that is, $S=\langle4,10,17\rangle$, then

```gap
s:=NumericalSemigroup(4,10,17);;
DeltaSet(s);
CatenaryDegree(s);
```

For example, if $m_1=2$, $m_2=3$, $b=1$, $c=2$, $a=3$, that is, $S=\langle6,8,9\rangle$, and we do the computations on $\texttt{GAP}$,

### 3. If $S$ has three Betti elements

If $\#\textrm{Betti}(S)=3$, the only cases in which  𝑆  has minimal catenary degree are the following.

#### 3.1. $r_{12}=r_{13}=1$, $r_{21}+r_{23}>c_2$ and  $c_1 \geq \max\{ r_{21}+r_{23}, r_{31}+r_{32} \}$

For example, if $S=\langle4,9,15\rangle$, we have $r_{12}=r_{13}=1$, $3+1=r_{21}+r_{23}>c_2=3$ and $6=c_1 > \max\{ r_{21}+r_{23}=3+1, r_{31}+r_{32}=3+2 \}=5$. We obtain

```gap
s:=NumericalSemigroup(4,9,15);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

Another example; if $S=\langle7,18,31\rangle$, we have $r_{12}=r_{13}=1$, $4+2=r_{21}+r_{23}>c_2=5$ and $7=c_1 = \max\{ r_{21}+r_{23}=4+2, r_{31}+r_{32}=3+4 \}=7$. We get

```gap
s:=NumericalSemigroup(7,18,31);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

```gap
BettiElements(s);
```

```gap
JupyterSplashDot(DotFactorizationGraph(Factorizations(49,s)));
```

Another example; if $S=\langle4,7,9\rangle$, we have $r_{12}=r_{13}=1$, $3+1=r_{21}+r_{23}>c_2=3$ and $4=c_1 = \max\{ r_{21}+r_{23}=3+1, r_{31}+r_{32}=1+2 \}=4$. Then,

```gap
s:=NumericalSemigroup(4,7,9);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

A last example; if $S=\langle6,13,17\rangle$, we have $r_{12}=r_{13}=1$, $3+2=r_{21}+r_{23}>c_2=4$ and $5=c_1 = \max\{ r_{21}+r_{23}=3+2, r_{31}+r_{32}=2+3 \}=5$. Then, we get

```gap
s:=NumericalSemigroup(6,13,17);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

#### 3.2. $r_{12}=r_{13}=1$, $r_{21}+r_{23}>c_2$, $r_{31}= r_{23}$ and $r_{21} \geq r_{32}$

It is a particular case of 3.1.

#### 3.3. $r_{12}=r_{13}=1$, $r_{21} = r_{32}$, $r_{21}+r_{23} > c_2$ and $r_{31}\geq r_{23}$

It is a particular case of 3.1.

#### 3.4. $c_3=2$, $r_{32}=r_{21}$ and $c_1 \geq c_2$

For example, if $S=\langle3,8,13\rangle$, we have $c_3=2$, $1=r_{32}=r_{21}=1$ and $7=c_1>c_2=2$. We get

```gap
s:=NumericalSemigroup(3,8,13);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

Another example; if $S=\langle4,5,7\rangle$, we have $c_3=2$, $2=r_{32}=r_{21}=2$ and $3=c_1=c_2=3$. Then, we obtain

```gap
s:=NumericalSemigroup(4,5,7);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

#### 3.5. $c_3=2$, $r_{31}=r_{12}$ and $c_2 \geq \max \{r_{21}+1, c_1\}$, excluding the case $c_2>c_1$ and $c_2 =r_{21}+1$

For example, if $S=\langle4,5,7\rangle$, we have $c_3=2$, $1=r_{31}=r_{12}=1$, $3=c_2 = \max \{r_{21}+1=2+1, c_1=3\}=3$. We get

```gap
s:=NumericalSemigroup(4,5,7);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

Another example; if $S=\langle10,11,37\rangle$, we have $c_3=2$, $3=r_{31}=r_{12}=3$, $7=c_2 = \max \{r_{21}+1=4+1, c_1=7\}=7$. Then,

```gap
s:=NumericalSemigroup(10,11,37);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

A last example; if $S=\langle9,10,34\rangle$, we have $c_3=2$, $2=r_{31}=r_{12}=2$ and $7=c_2 > \max \{r_{21}+1=4+1, c_1=6\}=6$. Thus,

```gap
s:=NumericalSemigroup(9,10,34);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

#### 3.6. $c_3=2$ and $r_{31}+r_{32}\geq \max \{c_1, c_2\}$

Note that we have some examples in 3.5. for the case $c_3=2$, $r_{31}+r_{32} \geq c_1$ and $r_{31}+r_{32}=c_2$. Let us search for examples with $c_3=2$, $r_{31}+r_{32} \geq c_1$ and $r_{31}+r_{32} > c_2$.

If $S=\langle3,7,11\rangle$, we have $c_3=2$, $5+1=r_{31}+r_{32} = c_1=6$ and $5+1=r_{31}+r_{32} > c_2=2$. Then, we get

```gap
s:=NumericalSemigroup(3,7,11);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```

As a last example, if $S=\langle7,11,38\rangle$. Then, $c_3=2$, $3+5=r_{31}+r_{32} > c_1=7$, $3+5=r_{31}+r_{32} > c_2=6$. We obtain

```gap
s:=NumericalSemigroup(7,11,38);;
DeltaSet(s);
CatenaryDegree(s);
MinimalPresentation(s);
```
